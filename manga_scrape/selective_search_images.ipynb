```python
import os
import random
import shutil
from PIL import Image
import cv2
import selective_search
from os import listdir
from os.path import isfile, join
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
komi_labeling = r'C:\Users\brunolopez\mldata\manga_scrape\komi_labeling'
```

```python
files = [komi_labeling + '/' + file for file in listdir(komi_labeling)]
```

```python
files_5 = files[:1]
files_5
```

```python
%matplotlib widget

for the_img in files_5:

    print(the_img)
    open_img = cv2.imread(the_img)
    boxes = selective_search.selective_search(open_img, mode='quality')

    # plt.imshow(open_img)
    # plt.imshow(boxes)
```

```python
boxes_filter = selective_search.box_filter(boxes, min_size=20, topN=80)
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(open_img)
for x1, y1, x2, y2 in boxes_filter:
    bbox = mpatches.Rectangle(
        (x1, y1), (x2-x1), (y2-y1), fill=False, edgecolor='red', linewidth=1)
    ax.add_patch(bbox)

plt.axis('off')
plt.show()
```
